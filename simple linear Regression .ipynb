 1. simple linear Regression 

```python
import pandas as pd 
df = pd.read_csv('ml_data_salary.csv')
df.head(3)
```

step-2 splitting data into traning data and testing data  

```python
X = df[['YearsExperience']]
y=df['Salary']
```

```python
X.head()
```

```python
y.head()
```

```python
#import library and split data
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state =0)
```

Step-3 Fit Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model = model.fit(X_train,y_train)
model 
```

Step-4  Poltting


```python
# train data plot
import matplotlib.pylab as plt 

plt.scatter(X_train,y_train)
```

```python
# test data plot 
import matplotlib.pylab as plt 

plt.scatter(X_test,y_test)
```

```python
#Train plot
import matplotlib.pylab as plt 

plt.scatter(X_train,y_train)
plt.plot(X_train,model.predict(X_train),color="Green")
plt.xlabel("Experince")
plt.ylabel("Salary")
plt.title("Train plot")
plt.show()
```

```python
#Test plot
import matplotlib.pylab as plt 

plt.scatter(X_test,y_test)
plt.plot(X_test,model.predict(X_test),color="red")
plt.xlabel("Experince")
plt.ylabel("Salary")
plt.title("Test plot")
plt.show()
```

### step-5 Testing or Evaluating your model 

```python
#Model Fitness
print('Score of Traning Data',model.score(X_train,y_train))
```

```python
#Model Fitness
print('Score of Testing Data',model.score(X_test,y_test) )
```
